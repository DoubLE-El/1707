```python
# turtle run

import turtle as t
import random

score = 0
playing = False

bad_turtle = t.Turtle()
bad_turtle.shape("turtle")
bad_turtle.color("red")
bad_turtle.speed(9)
bad_turtle.up()
bad_turtle.goto(0, 200)

food = t.Turtle()
food.shape("circle")
food.color("green")
food.speed(0)
food.up()
food.goto(0, -200)

def turn_right():
    t.setheading(0)

def turn_up():
    t.setheading(90)
    
def turn_left():
    t.setheading(180)
    
def turn_down():
    t.setheading(270)
    
def start():
    global playing
    if playing == False:
        playing = True
        t.clear()
        play()

def play():
    global score
    global playing
    t.forward(10)
    if random.randint(1, 5) == 3:
        angle = bad_turtle.towards(t.pos())
        bad_turtle.setheading(angle)
    speed = score + 5
    
    if speed > 15:
        speed = 15
    bad_turtle.forward(speed)
    
    if t.distance(bad_turtle) < 12:
        text = "Score: " + str(score)
        message("Game Over", text)
        playing = False
        score = 0
        
    if t.distance(food) < 12:
        score = score + 1
        t.write(score)
        cor_x = random.randint(-230, 230)
        cor_y = random.randint(-230, 230)
        food.goto(cor_x, cor_y)
    if playing:
        t.ontimer(play, 100)

def message(m1, m2):
    t.clear()
    t.goto(0, 100)
    t.write(m1, False, "center", 20)
    t.goto(0, -100)
    t.write(m2, False, "center", 15)
    t.home()

t.title("Turtle Run")
        
t.setup(500, 500)
t.bgcolor("yellow")
t.shape("turtle")
t.speed(0)
t.up()
t.color("black")
    
t.onkeypress(turn_right, "Right") 
t.onkeypress(turn_up, "Up")
t.onkeypress(turn_left, "Left")
t.onkeypress(turn_down, "Down")
t.onkeypress(start, "space")


t.listen()
message("Turtle Run", "[Press Space to Start]")

play()

t.mainloop()
```
